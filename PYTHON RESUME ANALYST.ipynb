```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
data = {

    "Candidate_Name":[
        "Arun",
        "Karthik",
        "Priya",
        "Rahul",
        "Sanjay"
    ],


    "Resume_Text":[

        "Python developer machine learning artificial intelligence NLP SQL data science",

        "Java developer spring boot software testing database management",

        "Python NLP deep learning tensorflow machine learning data analytics",

        "Frontend developer HTML CSS JavaScript React UI UX",

        "Python Django Flask SQL machine learning cloud AWS"
    ]

}


df = pd.DataFrame(data)

df
```

```python
def clean_text(text):

    # lowercase
    text = text.lower()


    # remove special characters
    text = re.sub('[^a-zA-Z]', ' ', text)


    # tokenization
    words = word_tokenize(text)


    # remove stopwords
    stop_words = set(stopwords.words('english'))


    filtered_words=[]


    for word in words:

        if word not in stop_words:
            filtered_words.append(word)


    return " ".join(filtered_words)
```

```python
df["Cleaned_Resume"] = df["Resume_Text"].apply(clean_text)

df
```

```python
job_description = """

We are looking for Python Developer
with Machine Learning, NLP,
Artificial Intelligence, SQL,
Deep Learning and Cloud skills.

"""
```

```python
clean_job = clean_text(job_description)


print(clean_job)
```

```python
documents = list(df["Cleaned_Resume"])


documents.append(clean_job)


documents
```

```python
tfidf = TfidfVectorizer()


tfidf_matrix = tfidf.fit_transform(documents)


print(tfidf_matrix.shape)
```

```python
resume_vectors = tfidf_matrix[:-1]


job_vector = tfidf_matrix[-1]


print(resume_vectors.shape)

print(job_vector.shape)
```

```python
similarity = cosine_similarity(
    resume_vectors,
    job_vector
)


similarity
```

```python
df["Match_Percentage"] = (

    similarity * 100

).round(1)


df
```

```python
ranking = df.sort_values(

    by="Match_Percentage",

    ascending=False

)


ranking
```

```python
for index,row in ranking.iterrows():

    print("Candidate Name :",row["Candidate_Name"])

    print("Resume Match :",row["Match_Percentage"],"%")

    print("----------------------------")
```
